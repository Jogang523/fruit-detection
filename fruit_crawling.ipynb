```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import json
```

```python
from urllib.request import urlretrieve
import os

os.mkdir('fruit')
```

```python

url = 'https://www.google.co.kr/?hl=ko'
driver = webdriver.Chrome()
driver.set_window_size(1000, 800)
driver.get(url)
```

```python
gsearch = input('검색어를 입력하세요: ')
driver.find_element(By.CSS_SELECTOR, '#APjFqb').send_keys(gsearch)
time.sleep(1)

search_btn = driver.find_element(By.CSS_SELECTOR, 'div.lJ9FBc > center > input.gNO89b')
search_btn.click()

image_btn = driver.find_element(By.LINK_TEXT, '이미지')
image_btn.click()

elems = driver.find_elements(By.CSS_SELECTOR, '#islrg > div.islrc > div:nth-child(2) > a.FRuiCf.islib.nfEiy > div.fR600b.islir > img')

SCROLL_PAUSE_TIME = 2

last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(SCROLL_PAUSE_TIME)

    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_height:
        try:
            time.sleep(SCROLL_PAUSE_TIME)
            driver.find_element(By.CSS_SELECTOR, "input.mye4qd").click()
        except:
            break
    last_height = new_height

len(elems)


for i, e in enumerate(elems[:500]):
    try:
        e.click()
        time.sleep(2)
        img_url =  driver.find_element(By.CSS_SELECTOR, '#Sva75c > div.A8mJGd.NDuZHe > div.LrPjRb > div.AQyBn > \
        div.tvh9oe.BIB1wf > c-wiz > div > div > div > div > div.v6bUne > div.p7sI2.PUxBg > a > img.sFlh5c.pT0Scc.iPVvYb').get_attribute('src')
        file_path = './.jpg'.format(i+1)
        urlretrieve(img_url, file_path)
    except:
        pass
```
