```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_data_importer = train_datagen.flow_from_directory(
    'C:/Users/M/jupyter-work/fruit/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data_importer = train_datagen.flow_from_directory(
    'C:/Users/M/jupyter-work/fruit/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
model = Sequential()
# l1
model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

#l2
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

#l3
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

#flat
model.add(Flatten())

#l4
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))

#ol
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os

model_dir = './model/'
if not os.path.exists(model_dir):
  os.mkdir(model_dir)

modelpath = './model/{epoch:02d}-{val_loss:.4f}.keras'
checkpoint = ModelCheckpoint(filepath=modelpath, monitor='val_loss', save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10)
```

```python
hist = model.fit(
    train_data_importer,
    epochs=50,
    validation_data=val_data_importer,
    callbacks=[checkpoint, early_stopping_callback]
)
```

```python
fig = plt.figure(figsize=(5,5))
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.legend(['train', 'val'])
plt.show()
```
